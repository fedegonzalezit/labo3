```python
# leo df_intermedio.parquet
import pandas as pd

df = pd.read_parquet('df_intermedio.parquet')
df
```

```python
# agrego columnas year, mes, quarter
df['year'] = df['fecha'].dt.year
df['mes'] = df['fecha'].dt.month
df['quarter'] = df['fecha'].dt.quarter

df
```

```python
#transformo columnas object a category y luego a int
object_cols = df.select_dtypes(include=['category']).columns
for col in object_cols:
    df[col] = df[col].cat.codes
df
```

```python
df.info()
```

```python
# creo nueva columna date_id que se basa en fecha pero es un uint8, 0 para la fecha mas vieja y suma de a 1 por mes
unique_dates = df['fecha'].drop_duplicates().sort_values()
date_to_id = {date: idx for idx, date in enumerate(unique_dates)}
df['date_id'] = df['fecha'].map(date_to_id).astype('uint8')
df['date_id'] = df['date_id'].astype('uint8')
```

```python
df[df["date_id"]==33]
```

```python
# elimino columna fecha
df = df.drop(columns=['fecha', "periodo"], errors='ignore')
```

```python
# veo cuantos nans hay
# hago fillna con 0
df = df.fillna(0)
df
```

```python
#agrego columna weight = 1
df['weight'] = 1

# agrupo por product_id con el siguiente criterio:
# cust_request_qty sum
# cust_request_tn sum
# tn sum
# stock_final max
# cat1, cat2, cat3 y brand, first
# sku_size, max
# year, mes quarter, date_id, weight, first
grouped_df = df.groupby(["date_id",'product_id']).agg({
    'cust_request_qty': 'sum',
    'cust_request_tn': 'sum',
    'tn': 'sum',
    'stock_final': 'max',
    'cat1': 'first',
    'cat2': 'first',
    'cat3': 'first',
    'brand': 'first',
    'sku_size': 'max',
    'year': 'first',
    'mes': 'first',
    'quarter': 'first',
    'weight': 'first',
}).reset_index()
```

```python
grouped_df
```

```python
grouped_df['cust_request_tn_diff'] = grouped_df['cust_request_tn'] - grouped_df['tn']
```

```python
import numpy as np
grouped_df['tn'] = grouped_df['tn'].apply(lambda x: np.log1p(x))
```

```python
# le hago a tn transformacion log1p agrupada por product_id
#grouped_df['tn'] = grouped_df.groupby('product_id')['tn'].transform(lambda x: pd.Series(np.log1p(x)))
grouped_df
```

```python
# nueva featurer que es cust_request_tn - tn y elimino cust_request_tn
grouped_df = grouped_df.drop(columns=['cust_request_tn'], errors='ignore')
# para cada product_id, se genera una serie, esa serie le aplico un scaler entre -1 y 1
from sklearn.preprocessing import MinMaxScaler, RobustScaler


#scalo todas las columnas menos product_id, weight, y date_id
#cols_to_scale = grouped_df.columns.difference(['product_id', 'weight', 'date_id', 'target', "sku_size", 'year', 'mes', 'quarter', 'cat1', 'cat2', 'cat3', 'brand'])
#scalers = {}
#for sub_serie in grouped_df.groupby('product_id')[cols_to_scale]:
#    for col in cols_to_scale:
#        scaler_idx = (sub_serie[0], col)
#        if col not in scalers:
#            scalers[scaler_idx] = RobustScaler()
#        grouped_df.loc[grouped_df['product_id'] == sub_serie[0], col] = scalers[scaler_idx].fit_transform(sub_serie[1][col].values.reshape(-1, 1))
grouped_df
```

```python
#creo 20 lags de las columnas cols_to_scale para las subseries de product_id
lags = 20
cols_to_lag = ["cust_request_qty", "tn", "stock_final", "sku_size", "cust_request_tn_diff"]
for col in cols_to_lag:
    for lag in range(1, lags + 1):
        grouped_df[f'{col}_lag_{lag}'] = grouped_df.groupby('product_id')[col].shift(lag)
grouped_df
```

```python
diffs = 12
for col in cols_to_lag:
    for diff in range(1, diffs + 1):
        grouped_df[f'{col}_diff_{diff}'] = grouped_df.groupby('product_id')[col].diff(diff)


means = [3, 6, 12]
for col in cols_to_lag:
    for mean in means:
        grouped_df[f'{col}_mean_{mean}'] = grouped_df.groupby('product_id')[col].transform(lambda x: x.rolling(window=mean, min_periods=1).mean())

```

```python
# vuelvo a calcular la columna target
grouped_df['target'] = grouped_df.groupby('product_id')['tn'].shift(-2)
#grouped_df["target"] = grouped_df['target'] - grouped_df["tn"]
# grouped_df['target_1'] = grouped_df.groupby('product_id')['tn'].shift(-1)
# grouped_df['target_2'] = grouped_df.groupby('product_id')['tn'].shift(-2)
#grouped_df['target_2'] = grouped_df['target_2'] - grouped_df['target_1']
#grouped_df['target_1'] = grouped_df['target_1'] - grouped_df['tn']
#grouped_df['target_1'] = grouped_df['target_1'].fillna(0)
#grouped_df['target_2'] = grouped_df['target_2'].fillna(0)
#grouped_df
```

```python
# separo el dataset, entreno hasta fecha = 2019-09 y testeo desde 2019-10
df_train = grouped_df[grouped_df['date_id'] <= 31]
df_test = grouped_df[grouped_df['date_id'] == 33]

#df_train = grouped_df[grouped_df['date_id'] <= 33]

#df_test = grouped_df[grouped_df['date_id'] == 35]
```

```python
# guardo df_trian y df_test como df_base_train.parquet y df_base_valid.parquet
df_train.to_parquet('df_base_train.parquet', index=False)
df_test.to_parquet('df_base_valid.parquet', index=False)
```

```python
# hago un lightgbm

import lightgbm as lgb
from sklearn.metrics import mean_squared_error

train_data = lgb.Dataset(df_train.drop(columns=['weight', "target"]),
                         label=df_train['target'])
valid_data = lgb.Dataset(df_test.drop(columns=["target", 'weight']),
                         label=df_test['target'])
callbacks = [lgb.log_evaluation(period=10)]
model = lgb.train(
    params={
        "objective": "regression",
        #"objective": "tweedie",
        "metric": "l2",
        "verbosity": -1,
        #boosting_type": "dart",
        #"learning_rate": 0.03,
        "num_leaves": 31,
        "max_bin": 1024
    },
    train_set=train_data,
    valid_sets=[valid_data],
    callbacks=callbacks,
    num_boost_round=200,
)


```

```python
preds = model.predict(df_test.drop(columns=['target','weight', "preds"], errors='ignore'))

df_test['preds'] = preds
```
